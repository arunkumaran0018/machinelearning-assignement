```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv(r"C:\Users\Arun Kumaran\Desktop\Renewable_Energy_Forecasting.csv")
features = data.select_dtypes(include=[float, int])
```

```python
scaler = StandardScaler()
features_scaled = scaler.fit_transform(features)
features_scaled 
```

```python
n_clusters = 3  
kmeans = KMeans(n_clusters=n_clusters, random_state=42)
cluster_labels = kmeans.fit_predict(features_scaled)
data['Cluster_Labels'] = cluster_labels
data['Cluster_Labels']
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(features_scaled[:, 0], features_scaled[:, 1], c=cluster_labels, cmap='viridis', s=50)
plt.title('K-Means Clustering Visualization')
plt.xlabel('Feature 1 (Scaled)')
plt.ylabel('Feature 2 (Scaled)')
plt.colorbar(label='Cluster')
plt.grid()
plt.show()
```

```python
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv(r"C:\Users\Arun Kumaran\Desktop\Renewable_Energy_Forecasting.csv")
features = data.select_dtypes(include=[float, int]) 

scaler = StandardScaler()
features_scaled = scaler.fit_transform(features)
```

```python
pca = PCA(n_components=2)
features_pca = pca.fit_transform(features_scaled)
```

```python
plt.figure(figsize=(10, 6))
# Assuming cluster_labels are obtained from hierarchical clustering as shown earlier
plt.scatter(features_pca[:, 0], features_pca[:, 1], cmap='blues', alpha=0.7)
plt.title('PCA Visualization of Clusters')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.grid()
plt.show()
```
